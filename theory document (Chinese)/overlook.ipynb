# 有限元理论概述

有限元法究竟是什么呢？从不同的角度自然有不同的回答。一些教材介绍了有限元软件的使用，其中也渗透着一些有限元法的原理。这些教材帮助非力学专业的学生和工程师对工程结构进行分析。他们使用强大的工作站运算着庞大规模的有限元问题，指导着工程实践，但许多人甚至不知道有限元法为什么要划分网格。一些教材以非常严密的数学推导建立了有限元法的理论基础，这使得人们可以放心地使用有限元法。但是这种放心在理论与工程之间存在着巨大的沟壑。事实上，只有若干最简单的偏微分方程和有限元离散方法被证明有所谓适定性和收敛性。大部分有限元计算只是在一种实用主义的信念下进行的，这当然是应当的，工程发展不会因为理论尚未支持就固步自封。自有限元法的建立开始，方法的提出总是远远早于方法的证明，后者甚至是无望的。那么作为一个计算力学的有限元法，应该如何理解呢？

有限元法和其他计算方法有什么区别呢？在我看来，有限元法给出共轭物理量，即给定作为自变量的每一单元的节点位移，有限元系统要给出每一单元的节点力，最后通过节点力的平衡反过来计算结点位移。
